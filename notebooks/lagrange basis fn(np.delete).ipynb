```python
def lagrange_basis(X, i, x):
    '''
    X = X nodes
    i = i_{th} Lagrange Basis
    x = coordinate at which the i_{th}
        Lagrange polynomials are to be evaluated.
    '''
    l = 1

    
    l = np.product((x-np.delete(X,i))/(np.sum(X[i])-np.delete(X,i)))      #use np.delete
    
    return l
                
```

```python

X = np.array([-1, -0.5, 0, 0.5, 1])
print(lagrange_basis(X, 2, 0.2))
```

```python
fig = plt.figure()
X = np.array([-1., -0.5, 0., 0.5, 1.])

for p in range (0,100):
    error = 0.
    r = random.uniform(-1.0, 1.0)
    f = 4*(r**4)-5*(r**2)+1
    error = abs(f-lagrange_basis(X, 2, r))
    pl.scatter(r, error)
pl.show()
```
