```python
def lagrange_basis(X, i, x):
    '''
    X = X nodes
    i = i_{th} Lagrange Basis
    x = coordinate at which the i_{th}
        Lagrange polynomials are to be evaluated.
    '''
    lx = 1.
    
    for m in range(0,X.size):
            if m != i:
                lx *= (x-X[m])/(X[i]-X[m])

    return lx
                
```
